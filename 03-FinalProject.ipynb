## Abstract

**The Big Idea: Do We Actually Hate Long Movies?**

With streaming apps giving us endless options, there is a constant debate about whether modern attention spans are shrinking and if long movies are losing their appeal. We wanted to see if "viewer fatigue" is real. Specifically, this study looks at whether longer runtimes cause casual viewers to leave worse reviews, while film critics who are used to slow-burn films remain unaffected. We also checked if this relationship changed across different decades or age ratings (like PG vs. R).


**How We Tested It:**

To figure this out, we built a dataset of 1,713 movies spanning a full century (1920 to 2021). We combined streaming data from Netflix, Hulu, Amazon Prime, and Disney+ with public review scores from IMDb (for audiences) and Rotten Tomatoes (for critics). Before analyzing anything, we cleaned up the data by turning messy runtime text into clean numbers and grouping mismatched age ratings into simple categories like G, PG, PG-13, and R.

**The Plot Twist: What the Data Showed**

Most movies in our data naturally clocked in between 80 and 120 minutes. But when we 
ran the numbers, we found something completely unexpected: opposing our initial guess, 
audiences and critics actually prefer longer movies. While the connection is weak, it 
is statistically positive for both groups:

- Audience Scores (IMDb): $r = 0.117$, $R^2 = 0.013$
- Critic Scores (Rotten Tomatoes): $r = 0.199$, $R^2 = 0.040$

Essentially, both groups respond to movie length in almost identical ways, completely 
debunking the myth that everyday audiences penalize long films.

**Real-World Limitations & Next Steps:**

This trend didn't change based on the era the movie was made, and age ratings had 
almost no impact on a movie's length. While these findings offer a clear 
counter-narrative to "viewer fatigue," our final report outlines critical limitations 
regarding modern streaming bias and details how future research can better track our 
evolving attention spans.

### Important Metrics
One of the most important variables in this dataset is runtime, which is measured in minutes. Runtime represents the total length of a movie and is the main variable used to test our hypothesis. Most movies in the dataset fall between around 80 and 120 minutes which is considered standard feature film length, although some are much shorter or longer. IMDb ratings are measured on a 0–10 scale and reflect audience opinions from users who leave ratings online. In general, movies with ratings above 7 are considered positively received by audiences, while ratings below 5 are usually viewed more negatively. Rotten Tomatoes scores are measured on a 0–100 scale and represent critic approval percentages. Higher Rotten Tomatoes scores generally indicate stronger critic reception.   


### Dataset Limitations
There are a few concerns with our dataset that could affect how well it tests our hypothesis. The biggest issue is that both rating columns come from self-selected groups rather than a true general audience. IMDB ratings come from users who chose to rate a film online, which skews younger, more male, and toward people who are already fans of the movie they are rating. People who started a long movie, got bored, and turned it off are exactly the viewers our hypothesis cares about, and they are the least likely to show up in the IMDB sample. This may bias the data against finding the effect we predicted. Rotten Tomatoes scores come from accredited critics, mostly covering US and UK theatrical releases, which is a different kind of bias and tends to favor prestige films that often happen to be longer. 
   
Another concern is that runtime in this dataset is confounded with content type. G rated films average 82 minutes while PG-13 films average 106 minutes, and our dataset includes 125 films under 60 minutes that are likely stand-up specials and short documentaries rather than feature films. This pulls the average runtime down to 98 minutes, which is shorter than what most people would consider a normal feature length, so our split between long and short movies may end up partly being a split between features and short-form content.

The dataset is also heavily skewed toward recent years, with about 67 percent of films from 2015 onward and only 121 films from before 1980, and the older films that did make the list are survivors that stayed culturally relevant, which inflates their ratings on both sides. We also do not have rating counts, so a film with 12 IMDB ratings is treated the same as one with 800,000. These biases are worth keeping in mind as we move into the analysis, since they could either mask or exaggerate the effect we are looking for.


### Load the Dataset

**Load the Dataset**
We initially had quite a few data sets that we were looking at. However since we are discussing the relationship between Movie runtimes, IMDB score, Rotten Tomato Score, the decade they were made, and their age ratings, there was no one data set that had all of these variables. To solve this problem we decided to combine all of these data sets.

Firstly we need to get all our data from each of the platforms into one big data set.

```python
import pandas as pd
import numpy as np

netflix = pd.read_csv('data/00-raw/netflix_titles.csv')
hulu = pd.read_csv('data/00-raw/hulu_titles.csv')
prime = pd.read_csv('data/00-raw/amazon_prime_titles.csv')
disney = pd.read_csv('data/00-raw/disney_plus_titles.csv')

# Combine
platforms = pd.concat([netflix, hulu, prime, disney], ignore_index=True)
```

Next We will filter the data and only keep the movie entries

```python
# Filter for Movies
movies_platforms = platforms[platforms['type'] == 'Movie'].copy()

movies_platforms = movies_platforms[['title', 'duration', 'rating', 'release_year']]
movies_platforms.rename(columns={'duration': 'Runtime', 'rating': 'Age_Rating_Platform', 'release_year': 'Year_Platform'}, inplace=True)

# Drop duplicates based on title
movies_platforms.drop_duplicates(subset=['title'], inplace=True)
```

Next we need to add the IMDB and Rotten tomato ratings to each movie

```python
# Load MoviesOnStreamingPlatforms
movies_sp = pd.read_csv('data/00-raw/MoviesOnStreamingPlatforms.csv')
movies_sp = movies_sp[['Title', 'Year', 'Age', 'Rotten Tomatoes']]

# Load IMDB
imdb = pd.read_csv('data/00-raw/imdb_ratings.csv')
imdb = imdb[['title', 'averageRating']]
imdb.rename(columns={'averageRating': 'IMDB_Rating', 'title': 'Title'}, inplace=True)

# Drop duplicates based on title for IMDB
imdb.drop_duplicates(subset=['Title'], inplace=True)

# Merge
merged = pd.merge(movies_sp, imdb, on='Title', how='inner')
merged = pd.merge(merged, movies_platforms, left_on='Title', right_on='title', how='inner')

# Combine Age and Year
merged['Final_Age'] = merged['Age'].fillna(merged['Age_Rating_Platform'])
merged['Final_Year'] = merged['Year'].fillna(merged['Year_Platform'])

# Select required columns
final_df = merged[['Title', 'IMDB_Rating', 'Rotten Tomatoes', 'Runtime', 'Final_Age', 'Final_Year']].copy()
final_df.rename(columns={'Final_Age': 'Age_Rating', 'Final_Year': 'Year'}, inplace=True)
```

Now that we have everything merged, we can clean our data by dropping any entries with null values

```python
final_df_clean = final_df.dropna(subset=['IMDB_Rating', 'Rotten Tomatoes', 'Runtime', 'Age_Rating', 'Year'])

print(f"Final dataset shape: {final_df_clean.shape}")
print(final_df_clean.head())

merged_data_path = 'data/01-interim/combined_movies_data.csv'
final_df_clean.to_csv(merged_data_path, index=False)
```

Now that everything is merged, there is still a few issues with our data set. Namely the ratings column is not standardized. In order to perform proper analysis, we need to standardize that column. Our mapping will be the following:

General Audiences
- 'G': 'G', 'TV-G': 'G', 'TV-Y': 'G', 'TV-Y7': 'G', 'TV-Y7-FV': 'G', 'all': 'G', 'ALL': 'G'

Parental Guidance Suggested
- 'PG': 'PG', 'TV-PG': 'PG', '7+': 'PG'

Teens
- 'PG-13': 'PG-13', 'TV-14': 'PG-13', '13+': 'PG-13', '16+': 'PG-13'

Restricted / Mature
- 'R': 'R', 'TV-MA': 'R', '18+': 'R', 'NC-17': 'R'

Not Rated
- 'NR': 'NR', 'UR': 'NR', 'UNRATED': 'NR', 'Not Rated': 'NR'

```python
# Load interim
df = pd.read_csv(merged_data_path)

rating_mapping = {
    'G': 'G', 'TV-G': 'G', 'TV-Y': 'G', 'TV-Y7': 'G', 'TV-Y7-FV': 'G', 'all': 'G', 'ALL': 'G',
    'PG': 'PG', 'TV-PG': 'PG', '7+': 'PG',
    'PG-13': 'PG-13', 'TV-14': 'PG-13', '13+': 'PG-13', '16+': 'PG-13',
    'R': 'R', 'TV-MA': 'R', '18+': 'R', 'NC-17': 'R',
    'NR': 'NR', 'UR': 'NR', 'UNRATED': 'NR', 'Not Rated': 'NR'
}

df['Age_Rating'] = df['Age_Rating'].replace(rating_mapping)

print("Standardized Age Ratings:")
print(df['Age_Rating'].value_counts())

Standardized_ratings = 'data/01-interim/combined_movies_data_standardized.csv'
df.to_csv(Standardized_ratings, index=False)
```

With our data set merged we can now finally load it

```python
df = pd.read_csv(Standardized_ratings)
display(df.head())
```

#### Section 2a Runtime vs IMDb Rating

In this section, we examine the relationship between movie runtime and IMDb ratings to test whether longer runtimes are associated with lower scores from general audiences. A scatter plot with a fitted regression line is used to visualize the overall trend, and the Pearson correlation coefficient is reported to quantify the strength and direction of the relationship.

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(10, 6))

sns.regplot(
    data=df,
    x='Runtime',
    y='IMDB_Rating',
    scatter_kws={'alpha': 0.4, 's': 25, 'color': '#4C72B0'},
    line_kws={'color': '#C44E52', 'linewidth': 2},
    lowess=False,
    ax=ax
)

ax.set_title('Runtime vs IMDb Rating', fontsize=14, fontweight='bold')
ax.set_xlabel('Runtime (minutes)')
ax.set_ylabel('IMDb Rating')
ax.grid(alpha=0.3)

corr = df['Runtime'].corr(df['IMDB_Rating'])
ax.text(0.05, 0.95, f'Pearson r = {corr:.3f}',
        transform=ax.transAxes, fontsize=11,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```

Runtime vs IMDb: Very weak positive correlation (r = 0.056), so runtime barely explains any variation in IMDb scores. Ratings span the full range across nearly all runtimes.

Distribution: Most films cluster between 80–130 minutes, with ratings packed between 5.0 and 7.5. A few outliers sit at very short (<20 min) and very long (>200 min) runtimes but don't pull the trend much.

Trend line: Nearly flat with a slight upward slope, suggesting longer films get marginally higher IMDb scores, not lower.

Takeaway: Contradicts our hypothesis that longer runtimes lower general audience ratings. IMDb users don't penalize longer films, and if anything reward them slightly.

#### Section 2b - Does Runtime Affect Critic Scores? Runtime vs. Rotten Tomatoes Score

In this section, we examine the relationship between movie runtime and Rotten Tomatoes ratings to test whether longer runtimes are associated with higher scores from critics. A scatter plot with a fitted regression line is used to visualize the overall trend, and the Pearson correlation coefficient is reported to quantify the strength and direction of the relationship.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data/02-processed/combined_movies_data_processed.csv')

avg_runtime = df['Runtime'].mean()
corr = df['Runtime'].corr(df['Rotten Tomatoes'])

# Plot
sns.lmplot(data=df, x='Runtime', y='Rotten Tomatoes',
           scatter_kws={'alpha': 0.3, 's': 15},
           line_kws={'color': 'red', 'label': 'Trend Line'},
           height=6, aspect=1.5)

# Average runtime line
plt.axvline(avg_runtime, color='black', linewidth=1.5,
            label=f'Avg Runtime ({avg_runtime:.0f} min)')

# Average Rotten Tomatoes score above/below average runtime
below = df[df['Runtime'] <= avg_runtime]['Rotten Tomatoes'].mean()
above = df[df['Runtime'] > avg_runtime]['Rotten Tomatoes'].mean()
print(f"Avg RT score (≤{avg_runtime:.0f} min): {below:.1f}")
print(f"Avg RT score (>{avg_runtime:.0f} min): {above:.1f}")

# Correlation text
print(f"Correlation: {corr:.3f}")

plt.title('Runtime vs. Rotten Tomatoes Score')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Rotten Tomatoes Score')
plt.legend()
plt.tight_layout()
plt.show()
```

Runtime vs Rotten Tomatoes: Weak positive correlation (r = 0.247), meaning runtime weakly represents variation in Rotten Tomatoes scores.

Distribution: Most movies are distributed between 80-120 minutes, with an average runtime of 98 minutes. There are runtime outliers below 30 minutes and a couple above 200 minutes. Rotten Tomatoe score outlars exist below 20 percent and above 90 percent scores. 

Trend line: The trend line has a positive slope, which suggests that longer movies recieve higher scores than shorter movies. This is reinforced by the findings that the average RT score for movies before 98 minutes is 54% while the average RT score longer than 98  minutes is 59.4%. 

Takeaway: Initially, the positive trend line and higher average score for longer length films suggest that our hypothesis (that Rotten Tomatoes critics will score longer length films higher than shorter ones) is true. However, due to the weak positive correlation (r = 0.247), this suggests that runtime is not a significant factor in determining critic reviews.

#### Section 3 - How Do Ratings Vary Across Age Groups? Runtime, IMDb, and RT by Age Rating

We compare runtime, IMDb, and Rotten Tomatoes scores across age rating groups using boxplots to explore whether age rating influences movie length or reception.

```python
import seaborn as sns

order = ['G', 'PG', 'PG-13', 'R', 'NR']
df_age = df[df['Age_Rating'].isin(order)]

fig, axes = plt.subplots(1, 3, figsize=(16, 6))
fig.suptitle('Rating Trends Across Age Ratings', fontsize=14, fontweight='bold')

for ax, col in zip(axes, ['Runtime', 'IMDB_Rating', 'Rotten Tomatoes']):
    sns.boxplot(data=df_age, x='Age_Rating', y=col, order=order, ax=ax,
                hue='Age_Rating', legend=False,
                medianprops=dict(color='red', linewidth=1.5))
    ax.set_title(col, fontsize=11, fontweight='bold')
    ax.set_xlabel('Age Rating')
    ax.grid(axis='y')

plt.tight_layout()
plt.show()
```

Runtime: All age groups have similar median runtimes (~95 min). PG and R films have the most outliers, including some very long films and suspiciously short ones likely from non-feature content.

IMDb: Medians are consistent across all groups (~6.5–7.0). G-rated films are slightly higher and more consistent. NR films trend slightly lower.

Rotten Tomatoes: More spread out than IMDb across all groups, medians around 55–65. NR films are the exception, lower median (~45) and much tighter, suggesting critics are consistently lukewarm on unrated content.

Takeaway: Age rating alone is not a strong predictor of runtime or ratings, distributions largely overlap, which partially challenges our hypothesis.


#### Section 4 - How Have Movie Runtimes and Ratings Changed Over the Decades?

In this section we will compare the runtimes and ratings over decades. To do this we will first group movies into their respective decades then find their average runtimes and ratings from both rotten tomatoes and IMDB. Both relationships are statistically significant (IMDb: r = 0.117, p < 0.001; RT: r = 0.199, p < 0.001), indicating that longer runtimes are associated with marginally higher scores on both platforms. Contrary to our hypothesis, longer movies do not receive lower audience ratings  if anything, the opposite is true, though the effect is too small to be practically meaningful.

### Second Analysis: Does the Runtime-Score Relationship Change Across Decades?

We next test whether the relationship between runtime and scores differs depending on when a movie was released. Our EDA showed that average runtimes and ratings have shifted across decades, so here we run separate regressions of runtime on IMDb and Rotten Tomatoes scores for each decade and compare the slopes. We use an interaction term in a linear regression model to formally test whether decade moderates the runtime-score relationship.

```python
import statsmodels.formula.api as smf

df_feat['Decade'] = (df_feat['Year'] // 10 * 10).astype(str)

decade_counts = df_feat['Decade'].value_counts()
valid_decades = decade_counts[decade_counts >= 30].index
df_dec = df_feat[df_feat['Decade'].isin(valid_decades)].copy()

model_imdb = smf.ols('IMDB_Rating ~ Runtime * C(Decade)', data=df_dec).fit()
model_rt   = smf.ols('Q("Rotten Tomatoes") ~ Runtime * C(Decade)', data=df_dec).fit()

print("IMDb interaction model p-value (Runtime:Decade):")
print(model_imdb.pvalues.filter(like='Runtime:').to_string())
print(f"\nIMDb model R²: {model_imdb.rsquared:.4f}")

print("\nRT interaction model p-value (Runtime:Decade):")
print(model_rt.pvalues.filter(like='Runtime:').to_string())
print(f"\nRT model R²: {model_rt.rsquared:.4f}")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Runtime vs. Scores by Decade', fontsize=14, fontweight='bold')

for ax, col, label in zip(axes,
                           ['IMDB_Rating', 'Rotten Tomatoes'],
                           ['IMDb Rating (0–10)', 'Rotten Tomatoes (0–100)']):
    for decade, group in df_dec.groupby('Decade'):
        if len(group) < 30:
            continue
        slope, intercept, r, p, _ = stats.linregress(group['Runtime'], group[col])
        x = np.linspace(group['Runtime'].min(), group['Runtime'].max(), 100)
        ax.plot(x, slope * x + intercept, label=decade)
    ax.set_xlabel('Runtime (minutes)')
    ax.set_ylabel(label)
    ax.legend(title='Decade', fontsize=8)
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('results/runtime_vs_scores_by_decade.png', dpi=150)
plt.show()
```

None of the Runtime:Decade interaction terms are statistically significant (all p > 0.05), indicating that decade does not meaningfully change how runtime relates to scores. While the plot shows some variation in slope steepness, older decades like the 1960s show a sharper positive slope, these differences are not statistically reliable given the small sample sizes in those decades. The runtime-score relationship is weak and positive regardless of era, which contradicts our hypothesis that post-2000 movies would show a stronger negative effect of runtime on ratings.

## Discusison and Conclusion

This exploratory data analysis analyzes how movie runtime affects critic scores and audience scores independently and whether this relationship differs by age rating and decade. The EDA was performed using movie runtime data from 1,713 movies from Netflix, Hulu, Disney+, and Amazon Prime, with critic scores and audience scores gathered from Rotten Tomatoes and IMDb. The results of the EDA do not match our initial hypothesis that longer films would receive lower ratings from a general audience but would be unaffected by critics. The results showed that there was a positive association between scores and runtime, indicating that longer runtimes have slightly higher scores among both audiences and critics, contrary to our hypothesis. However, due to R^2 values less than 4%, using movie runtime as a predictor of movie rating is not useful, as the effect is too small to be meaningful. 

Our secondary hypotheses were addressed at a broader level, rather than directly tested. However, the results of the EDA also contradicted our 2nd initial hypothesis that shorter movie runtimes would result in higher movie ratings for younger audiences. By comparing runtime and scores across age rating categories, medians across all age ratings were similar, indicating that the relationship between runtime and the age rating of a movie cannot be used as a predictor alone. Lastly, we hypothesized that highly rated films before the 2000s would have longer runtimes than highly rated films after the 2000s due to a change in attention span. The results of the EDA show that runtimes skyrocketed in the 1960s, with a peak of almost 108 minutes. However, there seems to be a steady decline, where the average modern movie is under 100 minutes after the 2000s. In our second analysis, we add to this by answering whether the runtime-score relationship changes across decades. The analysis shows that the runtime-score relationship is weak and positive regardless of era, because the analysis indicated that the decade does not meaningfully change the relationship between runtime and score. This ultimately contradicts our hypothesis that post-2000 movies would show a stronger negative effect of runtime on ratings. 

When compared with the studies discussed in the background section, the results of our exploratory data analysis agree with the studies done by the University of Malta students, while it differs from results found by Carnegie Mellon students. The study done by students from the University of Malta found that the content/script of the movie was a stronger predictor of movie rating than age-rating, runtime, decade produced, etc. This agrees with our results, which found that movie runtime had little effect (less than 4%) on both audience and critic scores, which supports the study done by the University of Malta. However, our results disagree with the study done by Carnegie Mellon students, which found that age ratings and title affected audience scores depending on the movie's release decade. Our analysis found that the relationship between run-time and scores, depending on the decade, was weak, indicating that decade was not a good predictor of movie score, while their study found otherwise.

One notable contradiction that resulted from our data analysis is that not only did audience and critic scores increase with longer runtime instead of reducing like proposed in our hypothesis, but critic scores were more sensitive to longer runtime, represented by a steeper slope. A possible explanation for this is that critics may be more comfortable watching longer films as a part of their job and view movies with longer runtimes as having better productions. Another contradiction was found in our initial EDA compared with our first analysis, where runtime vs IMDb scores were negligble(r=0.056) while runtime vs Rotten Tomatoes was relatively strong(r = 0.247). Upon the first analysis, the IMBb correlation jumped to 0.117 while RT dropped to 0.199 after filtering movies under 60 minutes and over 240 minutes. This means that the audience runtime-score relationship was decreased while the RT runtime-score was inflated due to outliers that were present before being filtered out. Due to this, our R^2 correlation estimates should be interpreted with this in mind, as they differ before and after the runtime outliers are filtered. 

Several dataset limitations are worth revisiting in light of our results. Both rating sources are self-selected samples rather than general audiences. IMDb ratings come from users who choose to rate a movie online, which skews towards a younger male audience and towards people who are already fans of the movie. This means that viewers who may have gotten bored and turned off a movie are less likely to rate the movie and appear in the IMDb dataset. Further, Rotten Tomatoes scores come from critics who cover US and UK movies, which may favor prestige films that may be longer. The dataset is also skewed towards recent years, where 67 percent of films are made after 2015, and only 121 films on the dataset are before 1980. Further, there is no count of the number of ratings a movie has, so the dataset does not have normalized movie ratings. More data about the movie production may help answer our question, as it may provide more information about the factors that movies with longer run times have that may explain the positive runtime-score relationship. Lastly, there is data that suggests that runtimes are changing. In the 1940's, movies were much shorter, likely as a result of World War II, peaking in 1960 again and declining since the 2000s. This suggests that cultural context through the decades may have affected the movie's runtime-score relationship.

Overall, the results disagree with the initial idea that audiences would become bored or tired from a long movie, while critics would likely be more engaged with the movie regardless of movie runtime. If this idea were true, we would find that the runtime-score slope would be negative for IMDb ratings and zero for Rotten Tomatoes scores; however, we find that both movie rating platforms have a positive slope, and the significance of runtime alone is weak. Since runtime alone has little significance on the score of the movie, but we still see a positive slope for both audience and critic scores, it is possible that other factors related to movie runtime may be affecting the scores. For example, it is possible that longer films have bigger studios, more funding, and higher quality productions, which may lead to higher ratings. It is possible that further analysis and more data about the production of movies may better answer our questions.
