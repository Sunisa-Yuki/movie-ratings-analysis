## Research Question

Does the runtime of a movie affect the reviews of critics and the general audience differently? How?

## Background and Prior Work

### Introduction:
We chose our topic because we wanted to explore data on the entertainment industry. Is art truly subjective, or can we point to certain characteristics that show us when a show/movie will have a large rating gap between critics and viewers. By exploring media data we can see different factors that contribute to ratings and we may uncover traits that make the critics and viewers disagree. 

From the project the Stanford University students conducted, they only got a ~ 52% prediction accuracy. They mentioned this was due to the fact that people that leave movie comments usually rate in extremes. If the movie was a 7, they’d rate it a 3-4 to bring the average down and vice versa. 

In the project from the University of Malta, they used film script to predict audience ratings and were more successful. The model's prediction was off by 1 point of the actual rating 84% of the time, and the classifier placed the film in the correct bucket 83% of the time.

In the analysis on IMDb ratings conducted by Carnegie Mellon University students, one discovery made was that movie content ratings and movie title affected IMDb scores depending on the decade the movie was released and could be used to predict the scores of upcoming movies. Further, they investigate how release season correlates with IMDb scores and profits.

Though many pieces of work explore rating prediction using several methods, none explore the characteristics that make a movie polarizing. That is what we want to dive deeper into, what makes a movie’s rating ambiguous? And how do we quantify that?

### Explanation of work already done: 
A similar project was conducted by using movie reviews to predict the amount of stars it got. This relates to our project because it ties in personal feedback with overall rating. This project used Naive Bayes to rate the movies/shows and was from students at Stanford University.

There was another project done by Karl Cini and John Abela from the Department of Computer Information Systems, Faculty of ICT, University of Malta, Msida Campus, Msida, MSD 2080, Malta. They read in script data and used NLP (Natural Language Processing) to predict film audience ratings. By testing out several methods such as XGBoost and MLPS, they found that SVR performed the best on unseen data and was stable results wise. This relates to our project by using how certain film characteristics can influence audience reviews. These characteristics can have strong overlap with what film critics look for. The characteristics that lead to better ratings may be subconsciously affecting audiences to score certain ways.

The last relevant project discussed is IMDb Movie Analysis by Josh Braverman, Pranav Reddy, and Seonho Park, who are students at Carnegie Mellon University. By scraping data from IMDb’s webpage extracting 15 qualitative and quantitative variables to understand trends and  answer their research questions. From the data gathered, they discovered the impact of age ratings on potential success of movies, how movie titles have evolved over the decades, as well as the impact a movie's release date has on IMDb scores and profits. This study is directly relevant to our project as it uses information from IMDb to form conclusions about the impacts of movie characteristics, which is essentially what our project will be doing.

## Hypothesis


We predict that movies with runtimes that are longer than the average movie runtime will recieve lower ratings from a general audience while critic reviews will be unaffected. We believe this because movies with longer runtimes may lose the interest of a general audience, while critics are more likely to base their ratings on other aspects of the movie.

## Data

#### Ideal Dataset:
The ideal dataset would contain one row per movie and include the following variables:
- IMDb user score (0-10) and number of IMDb votes
- Rotten Tomatoes critic score (0-100) and number of critic reviews
- Genre(s), MPAA age rating, runtime, release year
- Budget and box office revenue
- Director, lead cast, production studio
- Country of origin, language
- Streaming platform availability

Ideally this would cover at least 5,000+ movies across a wide range of years, genres, and budgets to ensure sufficient vairation in the disagreement variable. Data would be collected via official APIs (IMDb, Rotten Tomatoes) or aggregated from reputable film databases, and stored in a structured CSV or relational database with one row per film and with a unique movie ID as the key.
