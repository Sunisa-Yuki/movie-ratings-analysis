### Dataset Description

### Important Metrics
One of the most important variables in this dataset is runtime, which is measured in minutes. Runtime represents the total length of a movie and is the main variable used to test our hypothesis. Most movies in the dataset fall between around 80 and 120 minutes which is considered standard feature film length, although some are much shorter or longer. IMDb ratings are measured on a 0–10 scale and reflect audience opinions from users who leave ratings online. In general, movies with ratings above 7 are considered positively received by audiences, while ratings below 5 are usually viewed more negatively. Rotten Tomatoes scores are measured on a 0–100 scale and represent critic approval percentages. Higher Rotten Tomatoes scores generally indicate stronger critic reception.   


### Dataset Limitations
There are a few concerns with our dataset that could affect how well it tests our hypothesis. The biggest issue is that both rating columns come from self-selected groups rather than a true general audience. IMDB ratings come from users who chose to rate a film online, which skews younger, more male, and toward people who are already fans of the movie they are rating. People who started a long movie, got bored, and turned it off are exactly the viewers our hypothesis cares about, and they are the least likely to show up in the IMDB sample. This may bias the data against finding the effect we predicted. Rotten Tomatoes scores come from accredited critics, mostly covering US and UK theatrical releases, which is a different kind of bias and tends to favor prestige films that often happen to be longer. 
   
Another concern is that runtime in this dataset is confounded with content type. G rated films average 82 minutes while PG-13 films average 106 minutes, and our dataset includes 125 films under 60 minutes that are likely stand-up specials and short documentaries rather than feature films. This pulls the average runtime down to 98 minutes, which is shorter than what most people would consider a normal feature length, so our split between long and short movies may end up partly being a split between features and short-form content.

The dataset is also heavily skewed toward recent years, with about 67 percent of films from 2015 onward and only 121 films from before 1980, and the older films that did make the list are survivors that stayed culturally relevant, which inflates their ratings on both sides. We also do not have rating counts, so a film with 12 IMDB ratings is treated the same as one with 800,000. These biases are worth keeping in mind as we move into the analysis, since they could either mask or exaggerate the effect we are looking for.


### Load the Dataset

**Load the Dataset**
We initially had quite a few data sets that we were looking at. However since we are discussing the relationship between Movie runtimes, IMDB score, Rotten Tomato Score, the decade they were made, and their age ratings, there was no one data set that had all of these variables. To solve this problem we decided to combine all of these data sets.

Firstly we need to get all our data from each of the platforms into one big data set.

```python
import pandas as pd
import numpy as np

netflix = pd.read_csv('data/00-raw/netflix_titles.csv')
hulu = pd.read_csv('data/00-raw/hulu_titles.csv')
prime = pd.read_csv('data/00-raw/amazon_prime_titles.csv')
disney = pd.read_csv('data/00-raw/disney_plus_titles.csv')

# Combine
platforms = pd.concat([netflix, hulu, prime, disney], ignore_index=True)
```

Next We will filter the data and only keep the movie entries

```python
# Filter for Movies
movies_platforms = platforms[platforms['type'] == 'Movie'].copy()

movies_platforms = movies_platforms[['title', 'duration', 'rating', 'release_year']]
movies_platforms.rename(columns={'duration': 'Runtime', 'rating': 'Age_Rating_Platform', 'release_year': 'Year_Platform'}, inplace=True)

# Drop duplicates based on title
movies_platforms.drop_duplicates(subset=['title'], inplace=True)
```

Next we need to add the IMDB and Rotten tomato ratings to each movie

```python
# Load MoviesOnStreamingPlatforms
movies_sp = pd.read_csv('data/00-raw/MoviesOnStreamingPlatforms.csv')
movies_sp = movies_sp[['Title', 'Year', 'Age', 'Rotten Tomatoes']]

# Load IMDB
imdb = pd.read_csv('data/00-raw/imdb_ratings.csv')
imdb = imdb[['title', 'averageRating']]
imdb.rename(columns={'averageRating': 'IMDB_Rating', 'title': 'Title'}, inplace=True)

# Drop duplicates based on title for IMDB
imdb.drop_duplicates(subset=['Title'], inplace=True)

# Merge
merged = pd.merge(movies_sp, imdb, on='Title', how='inner')
merged = pd.merge(merged, movies_platforms, left_on='Title', right_on='title', how='inner')

# Combine Age and Year
merged['Final_Age'] = merged['Age'].fillna(merged['Age_Rating_Platform'])
merged['Final_Year'] = merged['Year'].fillna(merged['Year_Platform'])

# Select required columns
final_df = merged[['Title', 'IMDB_Rating', 'Rotten Tomatoes', 'Runtime', 'Final_Age', 'Final_Year']].copy()
final_df.rename(columns={'Final_Age': 'Age_Rating', 'Final_Year': 'Year'}, inplace=True)
```

Now that we have everything merged, we can clean our data by dropping any entries with null values

```python
final_df_clean = final_df.dropna(subset=['IMDB_Rating', 'Rotten Tomatoes', 'Runtime', 'Age_Rating', 'Year'])

print(f"Final dataset shape: {final_df_clean.shape}")
print(final_df_clean.head())

merged_data_path = 'data/01-interim/combined_movies_data.csv'
final_df_clean.to_csv(merged_data_path, index=False)
```

Now that everything is merged, there is still a few issues with our data set. Namely the ratings column is not standardized. In order to perform proper analysis, we need to standardize that column. Our mapping will be the following:

General Audiences
- 'G': 'G', 'TV-G': 'G', 'TV-Y': 'G', 'TV-Y7': 'G', 'TV-Y7-FV': 'G', 'all': 'G', 'ALL': 'G'

Parental Guidance Suggested
- 'PG': 'PG', 'TV-PG': 'PG', '7+': 'PG'

Teens
- 'PG-13': 'PG-13', 'TV-14': 'PG-13', '13+': 'PG-13', '16+': 'PG-13'

Restricted / Mature
- 'R': 'R', 'TV-MA': 'R', '18+': 'R', 'NC-17': 'R'

Not Rated
- 'NR': 'NR', 'UR': 'NR', 'UNRATED': 'NR', 'Not Rated': 'NR'

```python
# Load interim
df = pd.read_csv(merged_data_path)

rating_mapping = {
    'G': 'G', 'TV-G': 'G', 'TV-Y': 'G', 'TV-Y7': 'G', 'TV-Y7-FV': 'G', 'all': 'G', 'ALL': 'G',
    'PG': 'PG', 'TV-PG': 'PG', '7+': 'PG',
    'PG-13': 'PG-13', 'TV-14': 'PG-13', '13+': 'PG-13', '16+': 'PG-13',
    'R': 'R', 'TV-MA': 'R', '18+': 'R', 'NC-17': 'R',
    'NR': 'NR', 'UR': 'NR', 'UNRATED': 'NR', 'Not Rated': 'NR'
}

df['Age_Rating'] = df['Age_Rating'].replace(rating_mapping)

print("Standardized Age Ratings:")
print(df['Age_Rating'].value_counts())

Standardized_ratings = 'data/01-interim/combined_movies_data_standardized.csv'
df.to_csv(Standardized_ratings, index=False)
```
